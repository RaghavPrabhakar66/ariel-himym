```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import KDTree
```

```python
# Generate 100 random (X, Y) coordinates
coordinates = np.random.rand(100, 2)

# Plot the coordinates
plt.scatter(coordinates[:, 0], coordinates[:, 1])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random 2D Coordinates')
plt.show()
```

```python
output = KDTree(coordinates)
```

```python
# Query the nearest neighbors for the first coordinate
distances, indices = output.query(coordinates[0], k=2)

# Plot all coordinates
plt.scatter(coordinates[:, 0], coordinates[:, 1], label='All Points')

# Highlight the query coordinate in black
plt.scatter(coordinates[0, 0], coordinates[0, 1], color='black', s=100, label='Query Point')

# Highlight its neighbors in red (excluding the query point itself)
neighbor_indices = indices[1:]  # skip the first index (the point itself)
plt.scatter(coordinates[neighbor_indices, 0], coordinates[neighbor_indices, 1], color='red', s=100, label='Neighbors')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Query Point and Neighbors')
plt.show()
```
